```python
%matplotlib inline
from matplotlib import pyplot as plt
import cv2 as cv

plt.rcParams["figure.figsize"] = (18, 12)
```

```python
face_cascade = cv.CascadeClassifier('C:\\Users\\SBD\\CV_Lab-master\\haarcascade_frontalface_default.xml')
img = cv.imread('C:\\Users\\SBD\\CV_Lab-master\\profile.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# cv2 đọc ảnh dạng BGR, matplotlib hiển thị dạng RGB nên
# chuyển đôỉ BGR2RGB để hiển thị được chính xác màu
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=10)

for (x, y, w, h) in faces:
    cv.rectangle(img, (x,y), (x+w,y+h), (255, 0, 0), 10)

plt.imshow(img)
crop_img = img[y:y+h, x:x+w].copy() 
cv.imshow("cropped", crop_img)
cv.waitKey(0)
cv.imwrite('01.png',crop_img)
plt.show()
```
